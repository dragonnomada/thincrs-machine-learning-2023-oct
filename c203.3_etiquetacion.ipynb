# Caso de Uso - Etiquetado de Valores desconocidos

```python
import pandas as pd

iris = pd.read_csv("datasets/iris.csv")

iris.info()
```

```python
X = iris[ ["sepal.length", "sepal.width", "petal.length", "petal.width"] ].values

y = iris["variety"].values

X.shape, y.shape
```

```python
from sklearn.decomposition import PCA

Xp = PCA(n_components=2).fit_transform(X)

Xp.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x=Xp[:, 0], y=Xp[:, 1], hue=y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x=Xp[:, 0], y=Xp[:, 1])
plt.show()
```

```python
from sklearn.cluster import KMeans

clu = KMeans(n_clusters=3)

clu.fit(Xp)

yp = clu.predict(Xp)

import random

for _ in range(10):
    i = random.randint(0, len(y) - 1)
    print(yp[i], y[i])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x=Xp[:, 0], y=Xp[:, 1], hue=yp)
plt.show()
```
