### **EJEMPLO:** Recibo de Compra

```python
productos = [
    (123, 'Coca Cola', 17.5), # -> producto -> (id, nombre, precio) -> 3-tupla
    (124, 'Gansito', 21.5),
    (142, 'Galletas Mar칤as', 14.5),
    (189, 'Nito', 16.99),
    (231, 'Leche Alpura', 21.3333333333333333),
    (355, 'Yoplait', 11.761),
]

recibo = ""

for id, nombre, precio in productos:
  # recibo += f"[{id}] {nombre} ${precio}\n"
  recibo += "[{}] {:<20s} ${:.2f}\n".format(id, nombre, precio)
  # recibo += "[{}] {:^20s} ${:.2f}\n".format(id, nombre, precio)
  # recibo += "[{}] {:>20s} ${:.2f}\n".format(id, nombre, precio)

print(recibo)
```

## Colecciones

Existen tres colecciones importantes:

* Listas - Almacenan datos din치micamente (se pueden extender o vaciar)
* Tuplas - Empaquetan valores en forma de una variable especial llamada `n-tupla`
* Diccionarios - Almacenan los valores ligados a una clave (se pueden extender o contraer)

### Listas

Permiten operar una colecci칩n datos con los siguientes m칠todos:

- `lista.append(elemento)` - Agrega un elemento al final
- `lista.insert(칤ndice, elemento)` - Agrega un elemento en el 칤ndice y recorre a la derecha los dem치s elementos
- `lista.pop()` - Quita el 칰ltimo elemento (puede fallar)
- `lista.pop(칤ndice)` - Quita el elemento del 칤ndice (puede fallar)
- `lista.remove(elemento)` - Quita el primer elemento id칠ntico (puede fallar)
- `lista.count(elemento)` - Determina un entero con el n칰mero de elementos repetidos o cero

```python
frutas = ["manzana", "pera", "kiwi"] # 0-manzana | 1-pera | 2-kiwi

frutas.append('mango') # 0-manzana | 1-pera | 2-kiwi | 3-mango
frutas.insert(2, 'guayaba') # 0-manzana | 1-pera | 2-guayaba | 3-kiwi | 4-mango

print(frutas)
```

```python
print(list(enumerate(frutas)))
```

```python
[f"{indice}-{fruta}" for indice, fruta in enumerate(frutas)]
```

### Tuplas (Valores empaquetados/acoplados)

Los valores acoplados son valores fusionados en una tupla o paquete de valores que agrupa dichos valores para transportarlos o almacenarlos en cantidad.

```python
persona1 = ('Ana Ming', 23, 56.7)
persona2 = ('Diego Green', 18, 87.5)
persona3 = 'Bety Sims', 24, 65.4

persona1, persona2, persona3
```

```python
# LISTA [ elemento1, elemento2, elemento3, ... ]
# TUPLA ( valor1, valor2, valor3, ... )

personas = [persona1, persona2, persona3, ('Paco White', 32, 98.7)]

personas
```

```python
nombre, edad, peso = persona1 # Desacoplar / Destuplar / Desempaquetar

print(f"La persona se llama {nombre}, tiene {edad} a침os y pesa {peso} kg")
```

## Funciones (Tareas / Operaciones / Actividades / M칠todos)

Las funciones son fragmentos de entrada-salida reutilizables que se pueden llamar con diferentes par치metros de entrada para provocar una salida, que si no est치 definida ser치 `None` por defecto.

```python
def suma(a, b):
  return a + b

suma(1, 2)
```

```python
suma(100, 200)
```

```python
def division(a, b):
  assert not b == 0
  return a / b

c = division(1, 4)

c
```

```python
def descargar_imagen(url):
  import requests
  response = requests.get(url)
  assert response.ok
  imagen = response.content
  return imagen

batman_img = descargar_imagen("https://m.media-amazon.com/images/I/518YabTa8YL._AC_UF894,1000_QL80_.jpg")

with open("batman.jpg", "wb") as file:
  file.write(batman_img)

len(batman_img)
```

```python
import matplotlib.pyplot as plt # Importaci칩n con alias "matplotlib.pyplot -> plt"
from PIL import Image

batman_img2 = Image.open('batman.jpg')

plt.imshow(batman_img2)
plt.show()
```

```python
def escribir_imagen(nombre, imagen):
  with open(nombre, "wb") as file:
    file.write(imagen)
    print(f"Imagen guardada: {nombre}")
    print(f"Se escribieron {len(imagen)} bytes")

escribir_imagen("batman_demo.jpg", batman_img)
```

```python
def mostrar_imagen(nombre):
  import matplotlib.pyplot as plt
  from PIL import Image

  imagen = Image.open(nombre)

  plt.imshow(imagen)
  plt.show()

mostrar_imagen("batman.jpg")
```
